### importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
fruits = pd.read_excel("C:/Users/user/file/fruits.xlsx")
fruits.tail(5)
```

```python
plt.scatter(x="Sphericity",y="Weight",data=fruits[fruits.labels=="Apple"],c="red")
plt.scatter(x="Sphericity",y="Weight",data=fruits[fruits.labels=="Orange"],c="orange")
plt.show()
```

### taking independent variables as X and dependent field as y

```python
X=fruits[["Sphericity","Weight"]]
y=fruits["labels"]
```

### spliting data into train and test

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10,stratify=y)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

### scatter plot of train and test 

```python
plt.scatter(x="Sphericity",y="Weight",data=X_train[y_train=="Apple"],color='red')
plt.scatter(x="Sphericity",y="Weight",data=X_train[y_train=="Orange"],color='orange')
plt.scatter(x="Sphericity",y="Weight",data=X_test[y_test=="Apple"],color='red',marker="*")
plt.scatter(x="Sphericity",y="Weight",data=X_test[y_test=="Orange"],color='orange',marker="*")
plt.show()
```

### Entropy is the parameter to take a best feature as main node to start a tree.

```python
clf = DecisionTreeClassifier(criterion="entropy")
clf.fit(X_train,y_train)
```

```python
fig,ax = plt.subplots(figsize=[6,8])
_ = plot_tree(clf,feature_names=["Sphericity","Weight"],filled=True)
```

```python
print("train_score:",clf.score(X_train,y_train))
print("test_score:",clf.score(X_test,y_test))
confusion_matrix(y_test,clf.predict(X_test))
```

```python
clf = DecisionTreeClassifier(criterion="entropy")
clf.fit(X,y)
fig,ax = plt.subplots(figsize=[6,8])
_ = plot_tree(clf,feature_names=["Sphericity","Weight"],filled=True)
```

### Performance/accuracy of model

```python
print("score:",clf.score(X,y))
confusion_matrix(y,clf.predict(X))
```
