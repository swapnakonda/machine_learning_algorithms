### UNIVARIANT ANLYSIS(lets see with 1 feature can we clasify this data)

```python
df_setosa=df.loc[df["Species"]=="setosa"]
df_versicolor=df.loc[df["Species"]=="versicolor"]
df_verginica=df.loc[df["Species"]=="verginica"]
```

```python
### Univariant Diagram(lets take only sepallength)
plt.plot(df_setosa["Sepal.Length"],np.zeros_like(df_setosa["Sepal.Length"]),"o") 
plt.plot(df_versicolor["Sepal.Length"],np.zeros_like(df_versicolor["Sepal.Length"]),"*")
plt.plot(df_verginica["Sepal.Length"],np.zeros_like(df_verginica["Sepal.Length"]),"-")
plt.xlabel("petal_length")
plt.show()
#for y axis lets make everything zeros for this zero like taken nor of point i given same length as epal length
```

### bivariant analysis

```python
sns.FacetGrid(df,hue="Species",size=5).map(plt.scatter,"Petal.Length","Sepal.Width").add_legend();
plt.show()
```

```python
### from above we can see we can able to clasify setosa but virginica and versicolor has some point which are overlapping
### 
```
